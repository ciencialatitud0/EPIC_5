```python
# NumPy: Operaciones avanzadas de forma (Reshape, Concatenaci√≥n)

print("Ejemplos de manipulaci√≥n de la forma de arrays")

# Creaci√≥n de un array 1D
arr_1d = np.arange(12)
print("Array 1D original:", arr_1d)
# Reshape: Cambiar la forma de un array
# Convierte un array 1D de 12 elementos a un array 2D de 3 filas y 4 columnas
arr_2d = arr_1d.reshape((3, 4))
print("Array 2D (reshape a 3x4):\n", arr_2d)

# Tambi√©n podemos volver a un array 1D
arr_flat = arr_2d.flatten() # o arr_2d.ravel()
print("Array 1D (aplanado de nuevo):\n", arr_flat)
```

```python
print("\nEjemplos de Concatenaci√≥n de Arrays")

# Creamos dos arrays para concatenar
arr_a = np.array([[1, 2], [3, 4]])
arr_b = np.array([[5, 6], [7, 8]])

print("Array A:\n", arr_a)
print("Array B:\n", arr_b)

# Concatenaci√≥n vertical (por defecto axis=0)
# 'np.concatenate' es una funci√≥n general; 'np.vstack' es un caso especial para concatenaci√≥n vertical
concat_vertical = np.concatenate((arr_a, arr_b), axis=0) # Equivalente a np.vstack((arr_a, arr_b))
print("\nConcatenaci√≥n vertical (np.concatenate con axis=0):\n", concat_vertical)


# Concatenaci√≥n horizontal (axis=1)
# 'np.hstack' es un caso especial para concatenaci√≥n horizontal
concat_horizontal = np.concatenate((arr_a, arr_b), axis=1) # Equivalente a np.hstack((arr_a, arr_b))
print("Concatenaci√≥n horizontal (np.concatenate con axis=1):\n", concat_horizontal)

# Ejemplo de concatenar arrays de diferentes dimensiones (requiere ajuste si es necesario)
# Para concatenar arrays de diferentes dimensiones, a menudo se necesita usar np.newaxis
arr_c = np.array([10, 20, 30])
arr_d = arr_c[:, np.newaxis] # Transforma arr_c en una columna 2D
print("\nArray C (1D):", arr_c)
print("Array D (C como columna 2D):\n", arr_d)
print("Concatenaci√≥n horizontal de C y D (ejemplo de diferente dimensi√≥n, requiere cuidado):\n", np.concatenate((arr_c[:, np.newaxis], arr_d), axis=1))
```

```python
# Pandas: Indexaci√≥n expl√≠cita (.loc, .iloc) y Combinaci√≥n de DataFrames

print("Ejemplos de Indexaci√≥n con Pandas")

# Creamos un DataFrame de ejemplo
data_ciudades = {
    'Ciudad': ['Quito', 'Guayaquil', 'Cuenca', 'Ibarra', 'Loja'],
    'Pais': ['Ecuador', 'Ecuador', 'Ecuador', 'Ecuador', 'Ecuador'],
    'Poblacion': [3300000, 1600000, 800000, 700000, 350000],
    'Area_km2': [600, 100, 135, 140, 41]
}

df_ciudades = pd.DataFrame(data_ciudades, index=['UIO', 'GYE', 'CUE', 'IBA', 'LOJ'])
print("DataFrame de Ciudades:\n", df_ciudades)

```

```python
# .loc[]: Indexaci√≥n basada en etiquetas (labels)
print("\nUtilizando .loc[] (por etiquetas)")
# Seleccionar una fila por su etiqueta de √≠ndice
print("Fila con √≠ndice 'GYE':\n", df_ciudades.loc['GYE'])

# Seleccionar una columna por su etiqueta
print("\nColumna 'Poblacion':\n", df_ciudades.loc[:, 'Poblacion']) # el ':' significa todas las filas

# Seleccionar un valor espec√≠fico por fila y columna
print("\nPoblaci√≥n de Ibarra (IBA):", df_ciudades.loc['IBA', 'Poblacion'])

# Seleccionar un rango de filas y columnas por etiqueta
print("\nFilas de UIO a IBA y columnas Ciudad y Poblacion:\n", df_ciudades.loc['UIO':'IBA', ['Ciudad', 'Poblacion']])
```

```python
# .iloc[]: Indexaci√≥n basada en posiciones num√©ricas (integers)
print("\n--- Utilizando .iloc[] (por posiciones num√©ricas) ---")
# Seleccionar la primera fila (posici√≥n 0)
print("Primera fila (posici√≥n 0):\n", df_ciudades.iloc[0])

# Seleccionar una columna por su posici√≥n
print("\nSegunda columna (posici√≥n 1, 'Pais'):\n", df_ciudades.iloc[:, 1])

# Seleccionar un valor espec√≠fico por posici√≥n de fila y columna
print("\nPoblaci√≥n de la segunda ciudad (posici√≥n 1, columna 2):", df_ciudades.iloc[1, 2])

# Seleccionar un rango de filas y columnas por posici√≥n
print("\nPrimeras 2 filas y primeras 3 columnas:\n", df_ciudades.iloc[0:2, 0:3])
```

```python
print("\nEjemplos de Combinaci√≥n de DataFrames con pd.concat")

# Creamos dos DataFrames para combinar
df_paises_1 = pd.DataFrame({
    'Pais': ['Espa√±a', 'Francia'],
    'Capital': ['Madrid', 'Par√≠s']
})

df_paises_2 = pd.DataFrame({
    'Pais': ['Alemania', 'Italia'],
    'Capital': ['Berl√≠n', 'Roma']
})

print("DataFrame Paises 1:\n", df_paises_1)
print("DataFrame Paises 2:\n", df_paises_2)

# Concatenar verticalmente (por defecto, axis=0)
df_combinado_vertical = pd.concat([df_paises_1, df_paises_2], ignore_index=True)
print("\nDataFrame Combinado Verticalmente (con ignore_index=True):\n", df_combinado_vertical)

```

```python
print("\nEjemplo de Combinaci√≥n de DataFrames con pd.merge")

# Reset index to create a column to merge on (for demonstration)
df_paises_1_reset = df_paises_1.reset_index().rename(columns={'index': 'Original_Index'})

# Create df_info_extra with a corresponding index column
df_info_extra_with_key = pd.DataFrame({
    'Moneda': ['EUR', 'EUR', 'EUR', 'EUR'],
    'Original_Index': [0, 1, 0, 1] # Assuming a key to merge on
})

print("DataFrame Paises 1 (con √≠ndice reseteado):\n", df_paises_1_reset)
print("DataFrame Info Extra (con clave para merge):\n", df_info_extra_with_key)

# Merge the two DataFrames on the 'Original_Index' column
df_merged_example = pd.merge(df_paises_1_reset, df_info_extra_with_key, on='Original_Index', how='left')

print("\nDataFrame Combinado con pd.merge:\n", df_merged_example)
```

| Caracter√≠stica                   | **NumPy**                                               | **Pandas**                                                                   |
| -------------------------------- | ------------------------------------------------------- | ---------------------------------------------------------------------------- |
| üî¢ **Tipo de estructura**        | Arrays multidimensionales (`ndarray`)                   | Series (1D) y DataFrames (2D, tipo tabla)                                    |
| üß† **Prop√≥sito principal**       | C√°lculo num√©rico r√°pido, algebra lineal, matrices       | Manipulaci√≥n y an√°lisis de datos estructurados                               |
| ‚öôÔ∏è **Velocidad de ejecuci√≥n**    | Muy eficiente para operaciones num√©ricas intensivas     | M√°s vers√°til para manipular datos con etiquetas y columnas                   |
| üìä **Manejo de datos faltantes** | Limitado, no dise√±ado para `NaN` o `None`               | Muy robusto: permite filtrar, rellenar, eliminar valores nulos               |
| üß© **Indexaci√≥n**                | Por posici√≥n (√≠ndices num√©ricos)                        | Por etiquetas (columna, √≠ndice personalizado)                                |
| üîÑ **Funcionalidades comunes**   | Operaciones vectorizadas, funciones matem√°ticas         | Limpieza de datos, agrupaciones, fusiones, filtrado                          |
| üìà **Visualizaci√≥n**             | No tiene soporte integrado                              | Se complementa bien con `matplotlib` y `seaborn`                             |
| üì¶ **Ecosistema**                | Base para otras librer√≠as como `pandas`, `scikit-learn` | Se integra muy bien con `NumPy`, `matplotlib`, `statsmodels`, `scikit-learn` |


**NumPy** es ideal cuando necesitas trabajar con grandes cantidades de datos num√©ricos en forma de vectores o matrices, con alto rendimiento computacional.

**Pandas** est√° dise√±ado para el an√°lisis de datos estructurados (como archivos CSV, Excel o bases SQL), donde necesitas organizar, limpiar y explorar los datos de manera flexible.

## 5Ô∏è‚É£ Visualizaci√≥n avanzada con Seaborn y Matplotlib

```python
# Matplotlib: Estilo Orientado a Objetos (OO)
import matplotlib.pyplot as plt
print("Ejemplos de Matplotlib con Estilo Orientado a Objetos")

# Generar datos de ejemplo
# linspace es una funci√≥n de la librer√≠a NumPy que crea una secuencia de n√∫meros espaciados uniformemente dentro de un intervalo dado.
x = np.linspace(0, 10, 100)

# 1. Gr√°fico √∫nico con estilo OO
# plt.subplots() sin argumentos crea una figura y un solo eje (subplot)
fig, ax = plt.subplots(figsize=(8, 4)) # 'fig' es el objeto figura, 'ax' es el objeto ejes

# Plotear usando el m√©todo plot() del objeto 'ax'
ax.plot(x, np.sin(x), color='blue', linestyle='--', label='Seno(x)')

# Personalizar el gr√°fico usando los m√©todos del objeto 'ax'
ax.set_title('Funci√≥n Seno (Estilo Orientado a Objetos)')
ax.set_xlabel('Valores de X')
ax.set_ylabel('Valores de Y')
ax.legend()
ax.grid(True)
plt.show()
```

```python
# 2. M√∫ltiples subplots con estilo OO (como se menciona en el PDF)
# Creamos una figura y un array de objetos 'ax' (ej. 2 filas, 1 columna)
fig, ax = plt.subplots(2, 1, figsize=(8, 8), sharex=True) # sharex=True para compartir el eje X

# Accedemos a cada subplot a trav√©s del array 'ax'
ax[0].plot(x, np.sin(x), color='red')
ax[0].set_title('Gr√°fico Superior: Seno(x)')
ax[0].set_ylabel('Amplitud')
ax[0].grid(True)

ax[1].plot(x, np.cos(x), color='green')
ax[1].set_title('Gr√°fico Inferior: Coseno(x)')
ax[1].set_xlabel('Valores de X')
ax[1].set_ylabel('Amplitud')
ax[1].grid(True)

# Un t√≠tulo general para la figura
fig.suptitle('Comparaci√≥n de Funciones Seno y Coseno', fontsize=16)

# Ajustar el dise√±o para evitar superposiciones
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Ajusta el subt√≠tulo general
plt.show()
```

### üß™ Crear Columnas con `apply()` y Funciones Lambda/Definidas

```python
q1 = df_tips['tip'].quantile(0.25)
q3 = df_tips['tip'].quantile(0.75)

def clasificar_tip(x):
    if x < q1:
        return 'Baja'
    elif x > q3:
        return 'Alta'
    else:
        return 'Media'

df_tips['clasificacion_propina'] = df_tips['tip'].apply(clasificar_tip)
print("\nDataset con columna de clasificaci√≥n de propina:")
print(df_tips[['tip', 'clasificacion_propina']].head())
```

```python
resumen = df_tips.groupby(['day', 'clasificacion_propina'])['total_bill'].agg(['mean', 'count'])
resumen
```

```python
# Definir una funci√≥n m√°s compleja
def relacion_gasto_propina(row):
    return row['tip'] / row['total_bill']

# Aplicar la funci√≥n a cada fila usando axis=1
df_tips['ratio_propina_gasto'] = df_tips.apply(relacion_gasto_propina, axis=1)
print("\nDataset con ratio propina/gasto:")
print(df_tips[['total_bill', 'tip', 'ratio_propina_gasto']].head())
```

Usamos apply() para crear nuevas columnas basadas en condiciones o c√°lculos de otras columnas, tanto con lambdas como con funciones definidas.

## ‚ö° Data Scaling o Escalado de Datos

### üéØ Objetivos:
- Mejorar el rendimiento de los algoritmos de aprendizaje autom√°tico

Muchos algoritmos, como regresi√≥n log√≠stica, SVM, KNN o redes neuronales, son sensibles a la escala de los datos. Si las variables tienen diferentes unidades o rangos, el modelo puede aprender de forma ineficiente o err√≥nea.

- Acelerar la convergencia en algoritmos de optimizaci√≥n

Al escalar los datos, especialmente en modelos como redes neuronales o regresi√≥n lineal con gradiente descendente, se facilita que el proceso de entrenamiento sea m√°s r√°pido y estable, evitando oscilaciones o pasos desbalanceados.

- Evitar que las caracter√≠sticas con mayor escala dominen el modelo

El escalado garantiza que ninguna variable tenga m√°s peso que otras solo por su magnitud. As√≠ se asegura que todas las variables contribuyan equitativamente al modelo y se mantenga una representaci√≥n justa de los datos.

```python
# Importar las librer√≠as necesarias
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar el dataset Iris
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
df['species'] = df['target'].apply(lambda x: iris.target_names[x])
```

```python
# Mostrar las primeras filas del dataset
print("Dataset Iris original:")
display(df.head())
print("\nEstad√≠sticas descriptivas del dataset original:")
display(df.describe())
```

### ¬øCu√°ndo usar Normalizaci√≥n?

Transforma los datos para que todos est√©n entre un rango fijo, normalmente de 0 a 1.

‚úÖ Ventajas:
Muy √∫til cuando necesitas que todos los datos est√©n entre 0 y 1.

Bueno para algoritmos basados en distancias (k-NN, redes neuronales).

‚ö†Ô∏è Desventajas:
Muy sensible a valores at√≠picos (outliers). Si hay un n√∫mero muy grande o muy peque√±o, puede distorsionar todo el escalado.


---


- Algoritmos que dependen de distancias (como k‚ÄëNN, k‚ÄëMeans o redes neuronales), requieren que las caracter√≠sticas est√©n en una escala similar para evitar que las variables con magnitudes mayores dominen el c√°lculo de distancia.
- Si las caracter√≠sticas tienen rangos muy distintos (edad vs ingresos, por ejemplo), la normalizaci√≥n (Min‚ÄëMax) asegura que cada variable aporte proporcionalmente sin sesgos.
- Tambi√©n se recomienda cuando la distribuci√≥n es desconocida o no es Gaussiana, pues no requiere media ni varianza conocidas.



---



```python
# Seleccionar una caracter√≠stica para visualizar antes y despu√©s de la transformaci√≥n
feature_to_transform = 'sepal length (cm)'

#Aplicar Normalizaci√≥n (Min-Max Scaling)
print(f"\nAplicando Normalizaci√≥n (Min-Max Scaling) a '{feature_to_transform}'")
scaler_minmax = MinMaxScaler(feature_range=(0, 1))
df_normalized = df.copy()
df_normalized[feature_to_transform] = scaler_minmax.fit_transform(df[[feature_to_transform]])

print("\nEstad√≠sticas descriptivas despu√©s de Normalizaci√≥n:")
display(df_normalized.describe())
```

### ¬øCu√°ndo usar Estandarizaci√≥n?

Transforma los datos para que tengan media 0 y desviaci√≥n est√°ndar 1. Es decir, los convierte a una distribuci√≥n normal est√°ndar (tambi√©n se llama Z-score scaling).


‚úÖ Ventajas:
Bueno cuando los datos tienen una distribuci√≥n normal.

Menos afectado por outliers que Min-Max.

‚ö†Ô∏è Desventajas:
Si los datos no tienen una forma de campana (no normal), el resultado puede no tener mucho sentido.


---



- Ideal para algoritmos que asumen distribuci√≥n normal, como regresi√≥n log√≠stica, regresi√≥n lineal o PCA, donde se busca media cero y desviaci√≥n est√°ndar uno.
Funciona mejor si el conjunto de datos contiene outliers, ya que no se ve afectada por valores extremos como s√≠ pasa con la normalizaci√≥n.

- Tambi√©n se utiliza en modelos con regularizaci√≥n (Lasso, Ridge, SVM, redes neuronales), donde el algoritmo penaliza coeficientes en funci√≥n de su escala. Una estandarizaci√≥n adecuada mejora convergencia y estabilidad.

- Ayuda a acelerar el entrenamiento basado en descenso de gradiente al evitar que una caracter√≠stica domine las actualizaciones del gradiente.


---



```python
#Aplicar Estandarizaci√≥n (Standard Scaling)
print(f"\nAplicando Estandarizaci√≥n (Standard Scaling) a '{feature_to_transform}'")
scaler_standard = StandardScaler()
df_standardized = df.copy()
df_standardized[feature_to_transform] = scaler_standard.fit_transform(df[[feature_to_transform]])

print("\nEstad√≠sticas descriptivas despu√©s de Estandarizaci√≥n:")
display(df_standardized.describe())
```

```python
# Visualizaci√≥n de los resultados
plt.figure(figsize=(15, 6))

# Histograma de la caracter√≠stica original
plt.subplot(1, 3, 1)
sns.histplot(df[feature_to_transform], kde=True, color='skyblue')
plt.title(f"Original: {feature_to_transform}")
plt.xlabel(feature_to_transform)
plt.ylabel("Frecuencia")

# Histograma despu√©s de Normalizaci√≥n
plt.subplot(1, 3, 2)
sns.histplot(df_normalized[feature_to_transform], kde=True, color='lightgreen')
plt.title(f"Normalizado (0-1): {feature_to_transform}")
plt.xlabel("Valor Normalizado")
plt.ylabel("Frecuencia")

# Histograma despu√©s de Estandarizaci√≥n
plt.subplot(1, 3, 3)
sns.histplot(df_standardized[feature_to_transform], kde=True, color='salmon')
plt.title(f"Estandarizado (Media=0, Std=1): {feature_to_transform}")
plt.xlabel("Valor Estandarizado")
plt.ylabel("Frecuencia")

plt.tight_layout()
plt.show()
```

```python
#Comparaci√≥n de los valores de una peque√±a muestra
print("\nComparaci√≥n de valores de una muestra")
sample_indices = [0, 5, 10] # √çndices de ejemplo

print(f"\nOriginal '{feature_to_transform}':")
display(df.loc[sample_indices, [feature_to_transform]])

print(f"\nNormalizado '{feature_to_transform}':")
display(df_normalized.loc[sample_indices, [feature_to_transform]])

print(f"\nEstandarizado '{feature_to_transform}':")
display(df_standardized.loc[sample_indices, [feature_to_transform]])
```

```python
# Aplicar ambas transformaciones a todo el dataset para un uso posterior
# Normalizaci√≥n a todo el dataset
scaler_minmax_all = MinMaxScaler(feature_range=(0, 1))
df_minmax_scaled = pd.DataFrame(scaler_minmax_all.fit_transform(df.drop(['target', 'species'], axis=1)),
                                columns=df.drop(['target', 'species'], axis=1).columns)
df_minmax_scaled['target'] = df['target'] # Re-a√±adir el target
df_minmax_scaled['species'] = df['species'] # Re-a√±adir la especie

print("\nPrimeras 5 filas del dataset con todas las caracter√≠sticas normalizadas:")
display(df_minmax_scaled.head())
print("\nEstad√≠sticas descriptivas del dataset con todas las caracter√≠sticas normalizadas:")
display(df_minmax_scaled.describe())
```

```python
# Estandarizaci√≥n a todo el dataset
scaler_standard_all = StandardScaler()
df_standard_scaled = pd.DataFrame(scaler_standard_all.fit_transform(df.drop(['target', 'species'], axis=1)),
                                  columns=df.drop(['target', 'species'], axis=1).columns)
df_standard_scaled['target'] = df['target'] # Re-a√±adir el target
df_standard_scaled['species'] = df['species'] # Re-a√±adir la especie

print("\nPrimeras 5 filas del dataset con todas las caracter√≠sticas estandarizadas:")
display(df_standard_scaled.head())
print("\nEstad√≠sticas descriptivas del dataset con todas las caracter√≠sticas estandarizadas:")
display(df_standard_scaled.describe())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python

# Separar las caracter√≠sticas (X) y el objetivo (y)
X = df.drop(['target', 'species'], axis=1)
y = df['target']

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print(f"\nTama√±o del conjunto de entrenamiento: {X_train.shape}")
print(f"Tama√±o del conjunto de prueba: {X_test.shape}")
```

```python
# Normalizaci√≥n (Min-Max Scaling)
scaler_minmax = MinMaxScaler(feature_range=(0, 1))
X_train_normalized = scaler_minmax.fit_transform(X_train)
X_test_normalized = scaler_minmax.transform(X_test) # Usamos transform en el test set

print("Datos de entrenamiento normalizados (primeras 5 filas):\n", X_train_normalized[:5])
print("\nDatos de prueba normalizados (primeras 5 filas):\n", X_test_normalized[:5])

# Estandarizaci√≥n (Standard Scaling)
scaler_standard = StandardScaler()
X_train_standardized = scaler_standard.fit_transform(X_train)
X_test_standardized = scaler_standard.transform(X_test) # Usamos transform en el test set

print("\nDatos de entrenamiento estandarizados (primeras 5 filas):\n", X_train_standardized[:5])
print("\nDatos de prueba estandarizados (primeras 5 filas):\n", X_test_standardized[:5])
```

```python
# Modelo con datos originales (sin escalar)
print("\n Evaluando con Datos Originales")
model_original = LogisticRegression(max_iter=200, random_state=42) # max_iter para asegurar convergencia
model_original.fit(X_train, y_train)
y_pred_original = model_original.predict(X_test)
accuracy_original = accuracy_score(y_test, y_pred_original)
print(f"Precisi√≥n del modelo (Datos Originales): {accuracy_original:.4f}")
print("Reporte de Clasificaci√≥n (Datos Originales):\n", classification_report(y_test, y_pred_original))
```

```python
# Modelo con datos Normalizados
print("\nEvaluando con Datos Normalizados (Min-Max Scaling)")
model_normalized = LogisticRegression(max_iter=200, random_state=42)
model_normalized.fit(X_train_normalized, y_train)
y_pred_normalized = model_normalized.predict(X_test_normalized)
accuracy_normalized = accuracy_score(y_test, y_pred_normalized)
print(f"Precisi√≥n del modelo (Datos Normalizados): {accuracy_normalized:.4f}")
print("Reporte de Clasificaci√≥n (Datos Normalizados):\n", classification_report(y_test, y_pred_normalized))
```

```python
# Modelo con datos Estandarizados
print("\nEvaluando con Datos Estandarizados (Standard Scaling)")
model_standardized = LogisticRegression(max_iter=200, random_state=42)
model_standardized.fit(X_train_standardized, y_train)
y_pred_standardized = model_standardized.predict(X_test_standardized)
accuracy_standardized = accuracy_score(y_test, y_pred_standardized)
print(f"Precisi√≥n del modelo (Datos Estandarizados): {accuracy_standardized:.4f}")
print("Reporte de Clasificaci√≥n (Datos Estandarizados):\n", classification_report(y_test, y_pred_standardized))
```

```python
print("\nResumen de Precisi√≥n")
print(f"Precisi√≥n (Original): {accuracy_original:.4f}")
print(f"Precisi√≥n (Normalizado): {accuracy_normalized:.4f}")
print(f"Precisi√≥n (Estandarizado): {accuracy_standardized:.4f}")

# Visualizaci√≥n de la comparaci√≥n
metrics = ['Original', 'Normalizado', 'Estandarizado']
accuracies = [accuracy_original, accuracy_normalized, accuracy_standardized]

plt.figure(figsize=(8, 5))
sns.barplot(x=metrics, y=accuracies, hue=metrics, palette='viridis', legend=False)
plt.ylabel('Precisi√≥n del Modelo')
plt.title('Comparaci√≥n de Precisi√≥n con Diferentes Escalamientos')
plt.ylim(0.8, 1.0) # Ajustar los l√≠mites para ver mejor las diferencias si son peque√±as
plt.show()
```

##  üö® Datos Faltantes

### üéØ Objetivos:

En ciencia de datos, el manejo de datos faltantes es uno de los retos m√°s comunes al trabajar con informaci√≥n real. Los datos incompletos pueden aparecer por errores de captura, problemas de almacenamiento o porque simplemente ciertos valores no existen. Ignorar estos valores puede llevar a an√°lisis err√≥neos, modelos poco confiables o incluso fallos en los algoritmos. Por eso, es fundamental aprender a identificar, visualizar y tratar los datos faltantes de manera adecuada.

**¬øQu√© es un dato faltante? (`None`, `np.nan`, `pd.NA`)**
En Python existen varias formas de representar datos faltantes o valores nulos. Las m√°s comunes son:

- `None`: Objeto nulo de Python, indica ausencia de valor.
- `np.nan`: Valor especial de tipo flotante proporcionado por NumPy, representa "Not a Number".
- `pd.NA`: Introducido en versiones recientes de pandas, es una forma general de representar datos faltantes para distintos tipos de datos.

**Concepto y aplicaci√≥n de la matriz de sombras (shadow matrix)**

La **matriz de sombras** (o **shadow matrix**) es un concepto utilizado en el an√°lisis de datos y estad√≠sticas para representar la presencia o ausencia de datos en un conjunto de datos. Este tipo de matriz ayuda a visualizar y analizar los patrones de datos faltantes, facilitando el entendimiento de c√≥mo y por qu√© los datos faltan.

**Concepto de Matriz de Sombras**

La matriz de sombras es una representaci√≥n binaria del conjunto de datos original donde:

- **1** (o cualquier valor positivo) indica la presencia de un valor en esa posici√≥n del conjunto de datos original.

- **0** (o cualquier valor negativo) indica la ausencia de un valor en esa posici√≥n.

**Uso y Beneficios**

1. **Visualizaci√≥n de Datos Faltantes**:

- La matriz de sombras facilita la visualizaci√≥n de patrones en los datos faltantes. Puedes ver claramente qu√© filas o columnas tienen m√°s datos faltantes y si hay patrones espec√≠ficos.

2. **An√°lisis de Patrones de Datos Faltantes**:

- Ayuda a identificar si los datos faltantes est√°n distribuidos aleatoriamente o si siguen alg√∫n patr√≥n espec√≠fico.

3. **Preprocesamiento**:

- La matriz de sombras se utiliza en t√©cnicas de imputaci√≥n para entender c√≥mo los datos faltantes est√°n relacionados con otras variables y para mejorar los m√©todos de imputaci√≥n.

```python
(
    df_titanic
    #Genero una matriz de booleanos que me identifica con tru y false para que ,e puedan indicar cuales sno faltantes
    .isna()
    #Se remplaza para tener una matriz de datos con data m√°s informativa
    .replace({
        False: "Not missing",
        True: "Missing"
    })
    # se debe tener nombres coordinados entre las variables originales y estas que se crean para ello se agrega el sufijo _NA
    .add_suffix("_NA")
    #concatenamos la matriz original con esta matriz de sommbras que acabamos de crear
    .pipe(
        lambda shadow_matrix: pd.concat(
            [df, shadow_matrix],
            axis="columns"
        )
    )
)
```

**Automatizando la generaci√≥n de la matriz**

```python
df_titanic.missing.bind_shadow_matrix(only_missing = True) # only_missing = True toma en cuenta unicamente las variables que tienen valores faltantes
```

**Visualizaci√≥n de valores faltantes**

`missing_extension` puede hacer gr√°ficos para visualizar los patrones de ausencia en variables y casos.


```python
(df_titanic.missing.bind_shadow_matrix(only_missing = True).pipe(
        lambda df1:(sns.displot(
                data=df1,
                x="age",
                hue="pclass",
                kind="kde"
            )
        )
    )
)
```
